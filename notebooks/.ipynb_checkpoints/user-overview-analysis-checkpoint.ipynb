```python
def calculate_dispersion(x):
    return pd.Series({
        'Range': x.max() - x.min(),
        'IQR' : x.quantile(0.75) - x.quantile(0.25),
        'Variance': x.var(),
        'Standard Deviation': x.std()
    })
```

```python
dispersion_param = wrapper_df.df.select_dtypes(include=[np.number]).apply(calculate_dispersion)

print(dispersion_param)
```

```python
#import 
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
#converting to timestamp
wrapper_df.df['Start'] = pd.to_datetime(wrapper_df.df['Start'])
wrapper_df.df['End'] = pd.to_datetime(wrapper_df.df['End'])
```

```python
numeric_df = wrapper_df.df.select_dtypes(include=[np.number])
scaler = StandardScaler()
df_scaled = scaler.fit_transform(numeric_df)
```

```python
pca = PCA(n_components=2)
principal_components = pca.fit_transform(df_scaled)

df_pca = pd.DataFrame(data=principal_components, columns=['total_Email_bytes', 'total_Email_bytes'])
explained_variance = pca.explained_variance_ratio_
print(f"Explained vaiance: {explained_variance}")
```
